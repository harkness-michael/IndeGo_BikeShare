```python
# Function to categorize season based on month and day
def get_season(date):
    month = date.month
    day = date.day
    if (month == 3 and day >= 1) or (3 < month < 6) or (month == 5 and day <= 31):
        return 'spring'
    elif (month == 6 and day >= 1) or (6 < month < 9) or (month == 8 and day <= 31):
        return 'summer'
    elif (month == 9 and day >= 1) or (9 < month < 12) or (month == 11 and day <= 30):  # Adjusted for 30 days in November
        return 'fall'
    else:
        return 'winter'
```

```python
# Applying the function to create 'season' column
df['season'] = df['start_time'].apply(get_season)
```

```python
# Creating 'long/short' column based on 'duration'
df['long/short'] = df['duration'].apply(lambda x: 'short' if x < 15 else 'long')
```
