```python
#checking mixed types
for col in df.columns.tolist():
      weird = (df[[col]].map(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
      if len (df[weird]) > 0:
        print (col)
```

```python
#convert 'passholder_type' column to string data type
df_merged2['passholder_type'] = df_merged2['passholder_type'].astype(str)
```

```python
# Handle non-numeric values in 'bike_id'
df_merged2['bike_id'] = pd.to_numeric(df_merged2['bike_id'], errors='coerce').fillna(0).astype('int64')
```

```python
df_merged2['bike_id'].value_counts()
```

```python
#converting 'start_time' and 'end_time' to timestamps
df_merged2['start_time'] = pd.to_datetime(df_merged2['start_time'], format='%m/%d/%Y %H:%M')
```

```python
df_merged2['end_time'] = pd.to_datetime(df_merged2['end_time'], format='%m/%d/%Y %H:%M')
```

```python
# create 'trip_category_code' column with boolean data to use for analysis
df_merged2['trip_category_code'] = df_merged2['trip_route_category'].apply(lambda x: 0 if x == 'Round Trip' else 1)
```

```python
# create 'bike_type_code' column with boolean data for analysis
df_merged2['bike_type_code'] = df_merged2['bike_type'].apply(lambda x: 0 if x == 'standard' else 1)
```

```python
df_merged2.head(2)
```

```python
#creating 'date' and 'time_of_day' columns from 'start_time' and 'end_time'
df_merged2['start_date'] = df_merged2['start_time'].dt.date
```

```python
df_merged2['start_time_of_day'] = df_merged2['start_time'].dt.time
```

```python
df_merged2['end_date'] = df_merged2['end_time'].dt.date
```

```python
df_merged2['end_time_of_day'] = df_merged2['end_time'].dt.time
```
